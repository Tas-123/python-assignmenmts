```python
import pandas as pd
```

```python
data =pd.read_csv("/content/Pokemon.csv")
```

```python
data.head(n=10)
```

```python
################### HP HIGH LOW COLUMN ADDTION FUNCTION #######################


standard_Hp=60
def set_hp(val):
  if val < standard_Hp:
    return "HpLow"
  else:
    return "HpHigh"
```

```python
######################################## SPEED HIGH LOW COLUMN ADDITION FUNCTION ######################

standard_speed = data["Speed"].mean()

def set_speed(val):
  if val > standard_speed:
    return "SpeedLow"
  else:
    return "SpeedHigh"
```

```python
data['Speed_Type']=data['Speed'].apply(set_speed)
data['Hp_Type'] = data['HP'].apply(set_hp)
data
```
